```python
from dotenv import load_dotenv
import os

load_dotenv()
TMDB_API_KEY = os.getenv("TMDB_API_KEY")

if not TMDB_API_KEY:
    raise ValueError("You need to set TMDB_API_KEY in your .env file")
else:
    print("TMDB_API_KEY is set")
```

```python
import requests
movie_id = 238

# URL para la solicitud al API
url = f"https://api.themoviedb.org/3/movie/{movie_id}?api_key={TMDB_API_KEY}"

# Realizar la solicitud
response = requests.get(url)

# Verificar si la solicitud fue exitosa
if response.status_code == 200:
    # Convertir la respuesta JSON a un diccionario de Python
    movie_details = response.json()
    print(movie_details['original_title'])  # Aquí puedes manejar los detalles de la película como prefieras
    print('     ', movie_details['overview'])  # Aquí puedes manejar los detalles de la película como prefieras
else:
    print("Error retrieving data from TMDB API for movie:", movie_id)
```

```python
wikipedia_api_base = 'https://es.wikipedia.org/api/rest_v1/page/talk/'
page_titles = [
    'Star_Wars:_Episodio_IV_-_Una_nueva_esperanza',
    'Star_Wars:_Episodio_V_-_El_Imperio_contraataca',
    'Star_Wars:_Episode_VI_-_Return_of_the_Jedi',
    'Star_Wars:_Episodio_I_-_La_amenaza_fantasma',
    'Star_Wars:_Episodio_II_-_El_ataque_de_los_clones',
    'Star_Wars:_Episodio_III_-_La_venganza_de_los_Sith',
    'Star_Wars:_Episodio_VII_-_El_despertar_de_la_Fuerza',
    'Star_Wars:_Episodio_VIII_-_Los_últimos_Jedi',
    'Star_Wars:_Episodio_IX_-_El_ascenso_de_Skywalker',
    'Rogue_One:_una_historia_de_Star_Wars',
    'Han_Solo:_una_historia_de_Star_Wars'
]

import requests
import json
from bs4 import BeautifulSoup


def find_argument(content ):
    print(content)
    for topic in content.get('topics', []):
        if topic['html'] == 'Argumento':
            plot = []
            for reply in topic['replies']:
                html_plot = reply['html']
                plot.append(BeautifulSoup(html_plot, "html.parser").get_text())
            return '\n'.join(plot)
    return None

sw_data = []

for title in page_titles:
    url = f"{wikipedia_api_base}{title}"
    response = requests.get(url)
    if response.status_code == 200:
        page_content = response.json()
        plot = find_argument(page_content)
        t = title.replace('_', ' ') 
        print(t, plot)
        sw_data.append({'title': t, 'plot': plot})
    else:
        print(f"Error al realizar la solicitud para {title}:", response.status_code)

with open('star_wars_plots.json', 'w') as f:
    json.dump(sw_data, f, ensure_ascii=False, indent=4)

```
